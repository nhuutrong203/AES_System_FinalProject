```python
from datasets import load_dataset, concatenate_datasets

```

```python
datasets_list = [
    "nlpatunt/D_Ielts_Writing_Dataset",
    "TraTacXiMuoi/Ielts_writing_task1_academic",
    "chillies/IELTS_evaluations",
    "nlpatunt/D_Ielts_Writing_Task_2_Dataset",
    "chillies/IELTS-writing-task-2-evaluation"

]

for dataset_name in datasets_list:
    print(f"Processing: {dataset_name}")
    
    ds = load_dataset(dataset_name)
    all_splits = []
    
    for split in ds:
        temp = ds[split]
        
        if "image" in temp.column_names:
            temp = temp.remove_columns("image")
        
        all_splits.append(temp)
    
    merged_ds = concatenate_datasets(all_splits)
    
    filename = f"{dataset_name.replace('/', '_')}_all.csv"
    
    merged_ds.to_csv(filename)
    
    print(f"Saved: {filename}\n")

print("DONE ALL ")
```
