```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# =========================
# 1. Đường dẫn 3 file CSV
# =========================

files = {
    "ModernBert": "modernbert_model.csv",
    "MPNET": "mpnet.csv",
    "Roberta Large": "roberta_large_traininglogs.csv",
    "Roberta Base": "roberta_traininglogs.csv",
}

# Nếu file nằm trong folder khác thì sửa lại, ví dụ:
# "BERT Weak": "/content/drive/MyDrive/your_folder/training_logs_bert_weak.csv"


# =========================
# 2. Đọc dữ liệu
# =========================

logs = {}

for model_name, file_path in files.items():
    df = pd.read_csv(file_path)
    logs[model_name] = df

    print(f"\n{model_name}")
    print(df.head())



```

```python

# =========================
# 3. Plot Train Loss chung
# =========================

plt.figure(figsize=(8, 5))

for model_name, df in logs.items():
    plt.plot(
        df["epoch"],
        df["train_loss"],
        label=model_name
    )

plt.xlabel("Epoch")
plt.ylabel("Train Loss")
plt.title("Training Loss Comparison Across Models")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()

plt.show()
```

```python
# =========================
# 4. Plot Validation Loss chung
# =========================

plt.figure(figsize=(8, 5))

for model_name, df in logs.items():
    plt.plot(
        df["epoch"],
        df["val_loss"],
        label=model_name
    )

plt.xlabel("Epoch")
plt.ylabel("Validation Loss")
plt.title("Validation Loss Comparison Across Models")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()

plt.show()
```
