```python
df = df[df["Overall"] >= 3.0]

print("After removing band < 3:", df.shape)
```

```python
conflict_score = (
    df.groupby("Essay")["Overall"]
    .nunique()
)

bad_essay = conflict_score[conflict_score > 1].index

print("Essays with multiple scores:", len(bad_essay))

df = df[~df["Essay"].isin(bad_essay)]

print("After removing score conflict:", df.shape)

```

```python
conflict_q = (
    df.groupby("Essay")["Question"]
    .nunique()
)

bad_q = conflict_q[conflict_q > 1].index

print("Essays with multiple questions:", len(bad_q))

df = df[~df["Essay"].isin(bad_q)]

print("After removing multi-question:", df.shape)

```

```python
df["length"] = df["Essay"].str.split().str.len()

# IELTS realistic range
df = df[(df["length"] >= 80) & (df["length"] <= 600)]

print("After hard length filter (80-600):", df.shape)
```

```python
q1 = df["length"].quantile(0.25)
q3 = df["length"].quantile(0.75)
iqr = q3 - q1

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

print("IQR bounds:", lower, upper)

df = df[(df["length"] >= lower) & (df["length"] <= upper)]

print("After IQR filter:", df.shape)
```

```python
q1 = df["Overall"].quantile(0.25)
q3 = df["Overall"].quantile(0.75)
iqr = q3 - q1

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

print("IQR bounds:", lower, upper)
```

```python
def round_band(x):
    if pd.isna(x):
        return np.nan

    integer = int(x)
    decimal = x - integer

    if decimal < 0.25:
        return integer
    elif decimal <= 0.5:
        return integer + 0.5
    elif decimal < 0.75:
        return integer + 0.5
    else:
        return integer + 1
```

```python
df["criteria_mean"] = df[criteria].mean(axis=1).apply(round_band)

diff = (df["criteria_mean"] - df["Overall"]).abs()

print("Mean difference:", diff.mean())
print("Max difference:", diff.max())
```

```python
import re
```

```python
def basic_clean(text):
    text = str(text)
    text = text.lower()
    return text

df["clean_text"] = df["Essay"].apply(basic_clean)
```

```python
df["clean_text"]
```

```python
def tokenize(text):
    return re.findall(r'\b\w+\b', text)

df["tokens"] = df["clean_text"].apply(tokenize)
```

```python
df["word_count"] = df["tokens"].apply(len)

df["unique_words"] = df["tokens"].apply(lambda x: len(set(x)))

df["ttr"] = df["unique_words"] / df["word_count"]

df["avg_word_len"] = df["tokens"].apply(
    lambda x: np.mean([len(w) for w in x]) if len(x) > 0 else 0
)
```

```python
def split_sentences(text):
    return re.split(r'[.!?]+', text)

df["sentences"] = df["Essay"].apply(split_sentences)

df["sentence_count"] = df["sentences"].apply(lambda x: len([s for s in x if s.strip() != ""]))

df["avg_sentence_len"] = df.apply(
    lambda row: row["word_count"] / row["sentence_count"]
    if row["sentence_count"] > 0 else 0,
    axis=1
)

df["sentence_len_var"] = df["sentences"].apply(
    lambda sents: np.var([len(s.split()) for s in sents if s.strip() != ""])
    if len(sents) > 0 else 0
)
```

```python
df["long_word_ratio"] = df["tokens"].apply(
    lambda x: np.mean([len(w) > 6 for w in x]) if len(x) > 0 else 0
)

df["short_word_ratio"] = df["tokens"].apply(
    lambda x: np.mean([len(w) <= 3 for w in x]) if len(x) > 0 else 0
)
```

```python
def count_punctuation(text):
    return len(re.findall(r'[,.!?;:]', text))

df["punct_count"] = df["Essay"].apply(count_punctuation)

df["punct_density"] = df["punct_count"] / df["word_count"]
```
